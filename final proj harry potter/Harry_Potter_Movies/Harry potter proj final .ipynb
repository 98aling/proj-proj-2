```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
file_path = 'WorldCupMatches.csv'
df = pd.read_csv(file_path)

# Create a new feature: Total Goals
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']

# Visualize the distribution of the new feature (Total Goals)
plt.figure(figsize=(10, 6))
sns.histplot(df['Total Goals'], bins=20, kde=True)
plt.title('Distribution of Total Goals in World Cup Matches')
plt.xlabel('Total Goals')
plt.ylabel('Frequency')
plt.show()

# Box plot to visualize the spread of Total Goals by Year
plt.figure(figsize=(12, 8))
sns.boxplot(x='Year', y='Total Goals', data=df)
plt.title('Box Plot of Total Goals Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Goals')
plt.show()

# Scatter plot to explore the relationship between Total Goals and Attendance
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total Goals', y='Attendance', data=df)
plt.title('Scatter Plot of Total Goals vs Attendance')
plt.xlabel('Total Goals')
plt.ylabel('Attendance')
plt.show()

# Pairplot to visualize relationships between multiple numerical columns including Total Goals
sns.pairplot(df[['Home Team Goals', 'Away Team Goals', 'Total Goals']])
plt.suptitle('Pairplot Including Total Goals', y=1.02)
plt.show()
```

```python
def visualize_distribution_numerical(dataframe, column):
    """Visualize the distribution of a numerical column."""
    plt.figure(figsize=(10, 6))
    sns.histplot(dataframe[column], bins=20, kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
def pairplot(dataframe, columns):
    """Pairplot to visualize relationships between multiple numerical columns."""
    sns.pairplot(dataframe[columns])
    plt.suptitle('Pairplot of Selected Numerical Columns', y=1.02)
    plt.show()
```
