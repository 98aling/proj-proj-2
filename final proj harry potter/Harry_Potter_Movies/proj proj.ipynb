```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df1 = pd.read_csv('Chapters.csv', encoding='latin-1')
df2 = pd.read_csv('WorldCupMatches.csv', encoding='latin-1')
```

```python
df1.head()
df2.head()
```

```python
# Display basic information about the DataFrame
print("Original DataFrame:")
print(df2.info())
```

```python
# Check for missing values
print(df1.isnull().sum())
print(df2.isnull().sum())

# Handle missing values (example: fill with mean)
df1. fillna(df1.mean(), inplace=True)
df2. fillna(df2.mean(), inplace=True)
```

```python
# Example: Remove duplicates
df1 = df1.drop_duplicates()
df2 = df2.drop_duplicates()
```

```python
# Display basic information about the DataFrame
print("Original DataFrame:")
print(df2.info())
```

```python
#vizualisations 
# Visualize the distribution of a numerical column (e.g., attendance)
plt.figure(figsize=(10, 6))
sns.histplot(df2['Attendance'], bins=20, kde=True)
plt.title('Distribution of Attendance')
plt.xlabel('Attendance')
plt.ylabel('Frequency')
plt.show()
# Scatter plot to explore the relationship between two numerical columns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Home Team Goals', y='Away Team Goals', data=df2)
plt.title('Scatter Plot of Home Team Goals vs Away Team Goals')
plt.show()
# Box plot to visualize the spread of a numerical column by a categorical column
plt.figure(figsize=(12, 8))
sns.boxplot(x='Year', y='Attendance', data=df2)
plt.title('Box Plot of Attendance Over the Years')
plt.xlabel('Year')
plt.ylabel('Attendance')
plt.show()
# Visualization 2: Scatter plot of World Cup match attendance
plt.scatter(df2['Year'], df2['Attendance'], color='green', alpha=0.7)
plt.title('World Cup Match Attendance Over Years')
plt.xlabel('Year')
plt.ylabel('Attendance')
plt.show()
# Visualization 1: Bar Chart - Distribution of Chapters per Movie
chapters_per_movie = df1.groupby('Movie ID')['Chapter ID'].count()

plt.figure(figsize=(10, 6))
chapters_per_movie.plot(kind='bar', color='skyblue')
plt.title('Distribution of Chapters per Movie')
plt.xlabel('Movie ID')
plt.ylabel('Number of Chapters')
plt.show()


```

```python

# Create a new feature: Total Goals
df2['Total Goals'] = df2['Home Team Goals'] + df2['Away Team Goals']

# Visualize the distribution of the new feature (Total Goals)
plt.figure(figsize=(10, 6))
sns.histplot(df2['Total Goals'], bins=20, kde=True)
plt.title('Distribution of Total Goals in World Cup Matches')
plt.xlabel('Total Goals')
plt.ylabel('Frequency')
plt.show()
```
