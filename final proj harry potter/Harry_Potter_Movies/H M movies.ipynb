```python
import pandas as pd

# Read the dataframes
df1 = pd.read_csv('Chapters.csv', encoding='latin-1')
df2 = pd.read_csv('WorldCupMatches.csv', encoding='latin-1')

# Find the common columns
common_columns = df1.columns.intersection(df2.columns).tolist()

# Display the common columns
print("Common Columns:", common_columns)
```

```python
# Print column names
print("Columns in df1:", df1.columns)
print("Columns in df2:", df2.columns)

# Print data types
print("Data types in df1:", df1.dtypes)
print("Data types in df2:", df2.dtypes)

# Check if dataframes are empty
print("Is df1 empty?", df1.empty)
print("Is df2 empty?", df2.empty)
```

```python
# Choose common columns with integer data types
common_columns_int = df1.select_dtypes(include='int64').columns.intersection(df2.select_dtypes(include='int64').columns).tolist()

# Display the common columns with integer data types
print("Common Columns with Integer Data Types:", common_columns_int)
```

```python
# Print the first few rows of each dataframe
print("First few rows in df1:")
print(df1.head())

print("\nFirst few rows in df2:")
print(df2.head())
```

```python
# Resetting indices to use default integer indices
df1_reset = df1.reset_index()
df2_reset = df2.reset_index()

# Merging dataframes on the default integer index
merged_df = pd.merge(df1_reset, df2_reset, how='inner', left_index=True, right_index=True)

# Display the merged dataframe
print(merged_df)

```

```python
# Check for missing values
print("Missing values in merged_df:")
print(merged_df.isnull().sum())

# Handle missing values (example: fill with mean)
merged_df.fillna(merged_df.mean(), inplace=True)
```

```python
# Select only numeric columns
numeric_columns = merged_df.select_dtypes(include='number').columns

# Fill missing values in numeric columns with the mean
merged_df[numeric_columns] = merged_df[numeric_columns].fillna(merged_df[numeric_columns].mean())
```

```python
# Print column names for both dataframes
print("Columns in df1:", df1.columns)
print("Columns in df2:", df2.columns)
```

```python
print("Unique values in df1['Chapter ID']:", df1['Chapter ID'].unique())
print("Unique values in df2['Year']:", df2['Year'].unique())
```
