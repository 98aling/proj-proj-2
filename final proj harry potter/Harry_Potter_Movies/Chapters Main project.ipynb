```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    "Character Name": ["Harry Potter", "Ron Weasley", "Hermione Granger", "Albus Dumbledore", "Rubeus Hagrid"],
    "House": ["Gryffindor", "Gryffindor", "Gryffindor", "Gryffindor", "Gryffindor"]
}

df = pd.DataFrame(data)

# Count the number of characters in each house
house_counts = df['House'].value_counts()

# Bar chart for house distribution
plt.figure(figsize=(10, 6))
house_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Characters in Each House')
plt.xlabel('House')
plt.ylabel('Number of Characters')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    "Character Name": ["Harry Potter", "Ron Weasley", "Hermione Granger", "Albus Dumbledore", "Rubeus Hagrid"],
    "House": ["Gryffindor", "Gryffindor", "Gryffindor", "Gryffindor", "Gryffindor"],
    "Wand Wood": ["Holly", "NaN", "Vine", "Elder", "Oak"]
}

df = pd.DataFrame(data)

# Visualization 1: Count plot for house distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='House', data=df, palette='pastel')
plt.title('Distribution of Characters Across Houses')
plt.xlabel('House')
plt.ylabel('Number of Characters')
plt.show()

# Visualization 2: Count plot for wand wood distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Wand Wood', data=df, palette='muted')
plt.title('Distribution of Wand Woods')
plt.xlabel('Wand Wood')
plt.ylabel('Number of Characters')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    "Character Name": ["Harry Potter", "Ron Weasley", "Hermione Granger", "Albus Dumbledore", "Rubeus Hagrid"],
    "Wand Core": ["Phoenix Feather", "NaN", "Dragon Heartstring", "Thestral Tail Hair", "NaN"]
}

df = pd.DataFrame(data)

# Bar plot for wand core distribution
plt.figure(figsize=(10, 6))
df['Wand Core'].value_counts().sort_values().plot(kind='barh', color='skyblue')
plt.title('Distribution of Wand Cores')
plt.xlabel('Number of Characters')
plt.ylabel('Wand Core')
plt.show()
```
